```python
train_1['type'] = 0
test_1['Response'] = 0
test_1['type'] = 1
master_1 = pd.concat([train_1,test_1], axis = 0)
master_1
```

```python
from sklearn.preprocessing import LabelEncoder as le
master_1[['Vehicle_Damage','Gender']] = master_1[['Vehicle_Damage','Gender']].apply(le().fit_transform)
master_1['Vehicle_Age'] = master_1['Vehicle_Age'].replace({'> 2 Years':2,'< 1 Year':0,'1-2 Year':1})
master_1
```

import seaborn as sns
sns.heatmap(master_1.corr())

```python
cols_cat = ['Gender','Driving_License','Previously_Insured','Vehicle_Damage','Vehicle_Age','Region_Code']
```

```python
cols_ignore = ['id','type']
cols_y = ['Response']
cols_X = list(set(master_1.columns) - set(cols_ignore) - set(cols_y))
```

```python
X = master_1[master_1['type']==0][cols_X]
X_test = master_1[master_1['type'] == 1][cols_X]
y = master_1[master_1['type']==0][cols_y]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X,y,test_size=0.4)
```

```python
import lightgbm as lgb
from lightgbm import LGBMClassifier
lgb = LGBMClassifier(n_estimators=55, num_leaves = 100, learning_rate = 0.08, max_depth = 50)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.metrics import roc_auc_score
sss = StratifiedShuffleSplit(n_splits=5, test_size=0.4, random_state=0)
n = 1
y_preds = []
for train_index, val_index in sss.split(X, y):
    print("TRAIN:", train_index, "VALIDATION:", val_index)
    X_train, X_val = X.iloc[train_index,:], X.iloc[val_index,:]
    y_train, y_val = y.iloc[train_index,:], y.iloc[val_index,:]
    clf = lgb.fit(X_train, y_train ,eval_metric='auc',eval_set=(X_val,y_val),verbose = 2,categorical_feature=cols_cat)
    print(str(n)+'. Train:',roc_auc_score(y_train,clf.predict_proba(X_train)[:,1]),'Validation:',roc_auc_score(y_val,clf.predict_proba(X_val)[:,1]))
    y_test = clf.predict_proba(X_test)[:, 1]
    y_preds.append(y_test)
    n = n+1
```

```python
y_preds_df = pd.DataFrame(y_preds).T
y_preds_df['Response'] = y_preds_df.mean(axis = 1)
y_preds_df
```

```python
sub_df = pd.concat([master_1[master_1['type'] == 1]['id'],y_preds_df['Response']], axis = 1)
#sub_df.columns = ['id','Response']
sub_csv = sub_df.to_csv('sub.csv', index = False)
sub_df
```
